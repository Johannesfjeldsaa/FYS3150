```python
xmin = 0
xmax = 1
xvals = np.linspace(xmin, xmax, 1000)
```

```python
def u(x, xrange=(0, 1)):
    # check if x value is within the domain
    if x < xrange[0] or x > xrange[-1]:
        raise ValueError('x is outside of xspace')
    # implement boundary conditions
    if x == 0 or x == 1:
        return 0
    else:
        return 1 - (1 - np.exp(-10)) * x - np.exp(-10 * x)
```

```python
plt.plot(xvals, [u(x) for x in xvals])
plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Analytical Solution')
plt.show()
```
